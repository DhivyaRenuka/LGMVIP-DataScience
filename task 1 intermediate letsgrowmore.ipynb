```python
#we only takes an important columns which is required only in our dataframe.
df1=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
df1.head()
```

```python
df1.info()
```

```python
profile= ProfileReport(df1, title="EDA", explorative=True)
```
